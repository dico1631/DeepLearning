# 1. alpha(open)

## 대문자 알파벳 세 개로 만들수 있는 단어를 출력한다.

- 입력
입력은 없다.

- 출력
사전 순으로 줄 바꿈없이 한 칸의 공백을 두고 출력한다.

- 입출력 예

입력: 없음

출력:
AAA AAB AAC ... ZZZ

````python
# 65: A, 90: Z
for i in range(ord('A'),ord('Z')):
    for j in range(ord('A'),ord('Z')):
        for k in range(ord('A'),ord('Z')):
            print(f'{chr(i)}{chr(j)}{chr(k)}', end=" ")
````

# 2. 마름모 구하기

1 ~ 15 사이의 홀수만입력됩니다. <br>
홀수 마름모를 그려주세요. 
--------------------------------------------------------------
- 입력:
5

- 출력:

```
  * 
 ***
*****
 ***
  *
  ```

````python
n = int(input())

# 꽉 찰때까지 1개씩 앞으로
# 꽉 찰때 = 0일때
start = n//2
start

# 꽉 찰때까지 1개씩 뒤로
# 꽉 찰때 = start가 0 일때
end = n//2
end

increase = True
increase
````

````python
increase = True
for i in range(n):
    for j in range(5):
        if j==start:
            print("*" * (end-start+1), end="")
            print(" " * (n-1-end), end="")
            if (start != 0) & (increase == True):
                start -= 1
                end += 1
            elif (start == 0) | (increase == False):
                increase = False
                start += 1
                end -= 1
            break
        else:
            print(" ", end="")
    print()
````

- 몽찡의 풀이

````python
num = 5
for i in range(1, num+1):
    print(abs(5-(2*i-1))//2)
````

````python
num = int(input("홀수를 넣어주세요: "))
for i in range(1,num+1):
    blank = abs((num-(2*i-1))//2)
    star = num-blank*2
    print(" "*blank, "*"*star, " "*blank)
````
